```python
import pandas as pd

df = pd.read_csv('nawy_imputition.csv')

#  Use .extract() with a Regex pattern to create 3 new columns
# The pattern looks for: [Text] + " in " + [Text] + " by " + [Text]
pattern = r'(?P<Project_Name>.*) in (?P<Location>.*) by (?P<Developer>.*)'
split_columns = df['Compound_Name'].str.extract(pattern)

# Add the new columns to your main dataframe
df = pd.concat([df, split_columns], axis=1)

# View the results
print(df[['Compound_Name', 'Project_Name', 'Location', 'Developer']].head())

# Optional: Save the new version to a CSV
df.to_csv('nawy_processed_splits.csv', index=False)
```
